```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mcol(v):
    return v.reshape((v.size, 1))
def load(fname):
    DList = []
    labelsList = []
    hLabels = {
        'False': 0,
        'True': 1
        }

    with open(fname) as f:
        for line in f:
            try:
                attrs = line.split(',')[0:-1]
                attrs = mcol(np.array([float(i) for i in attrs]))
                DList.append(attrs)
                labelsList.append(line.split(',')[-1].strip())
            except:
                print('Error in line: ', line)
                pass

    return np.hstack(DList), np.array(labelsList, dtype=np.int32)
```

```python
data, labels = load('./train.csv')
data.shape, labels.shape
```

```python
def plot_hist(feature, labels, bins=10):
    l = ["False", "True"]
    plt.hist(feature[labels == 0].flatten(), alpha=0.5, label=l[0], bins=bins, density=True)
    plt.hist(feature[labels == 1].flatten(), alpha=0.5, label=l[1], bins=bins, density=True)
    plt.legend()
    plt.show()
```

```python
for i in range(data.shape[0]):
    plot_hist(data[i], labels, bins=100)
```

```python
def plot_scatter(feat1, feat1_name, feat2, feat2_name, labels):
    l = ["False", "True"]
    plt.scatter(feat1[labels == 0], feat2[labels == 0], label=l[0])
    plt.scatter(feat1[labels == 1], feat2[labels == 1], label=l[1])
    plt.xlabel(feat1_name)
    plt.ylabel(feat2_name)
    plt.legend()
    plt.show()
```

```python
# plot scatter for all pairs of features
for i in range(data.shape[0]):
    for j in range(i+1, data.shape[0]):
        plot_scatter(data[i], f'feature {i}', data[j], f'feature {j}', labels)
```
