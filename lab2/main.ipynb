```python
import sklearn.datasets
def load_iris():
    return sklearn.datasets.load_iris()['data'].T, sklearn.datasets.load_iris()['target'], sklearn.datasets.load_iris()['feature_names']
```

```python
D, iris_labels, feature_names = load_iris()
```

```python
D.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_hist(feature, labels, bins=10):
    l = ["setosa", "versicolor", "virginica"]
    plt.hist(feature[labels == 0].flatten(), alpha=0.5, label=l[0], bins=bins, density=True)
    plt.hist(feature[labels == 1].flatten(), alpha=0.5, label=l[1], bins=bins, density=True)
    plt.hist(feature[labels == 2].flatten(), alpha=0.5, label=l[2], bins=bins, density=True)
    plt.legend()
    plt.show()
bins = 10
plot_hist(D[0], iris_labels, bins)
plot_hist(D[1], iris_labels, bins)
plot_hist(D[2], iris_labels, bins)
plot_hist(D[3], iris_labels, bins)
```

```python
def plot_scatter(feat1, feat1_name, feat2, feat2_name, labels):
    l = ["setosa", "versicolor", "virginica"]
    plt.scatter(feat1[labels == 0], feat2[labels == 0], label=l[0])
    plt.scatter(feat1[labels == 1], feat2[labels == 1], label=l[1])
    plt.scatter(feat1[labels == 2], feat2[labels == 2], label=l[2])
    plt.xlabel(feat1_name)
    plt.ylabel(feat2_name)
    plt.legend()
    plt.show()
plot_scatter(D[0], feature_names[0], D[1], feature_names[1], iris_labels)
plot_scatter(D[0], feature_names[0], D[2], feature_names[2], iris_labels)
plot_scatter(D[0], feature_names[0], D[3], feature_names[3], iris_labels)
plot_scatter(D[1], feature_names[1], D[0], feature_names[0], iris_labels)
plot_scatter(D[1], feature_names[1], D[2], feature_names[2], iris_labels)
plot_scatter(D[1], feature_names[1], D[3], feature_names[3], iris_labels)
plot_scatter(D[2], feature_names[2], D[0], feature_names[0], iris_labels)
plot_scatter(D[2], feature_names[2], D[1], feature_names[1], iris_labels)
plot_scatter(D[2], feature_names[2], D[3], feature_names[3], iris_labels)
plot_scatter(D[3], feature_names[3], D[0], feature_names[0], iris_labels)
plot_scatter(D[3], feature_names[3], D[1], feature_names[1], iris_labels)
plot_scatter(D[3], feature_names[3], D[2], feature_names[2], iris_labels)
```

```python
def vcol(v):
    return v.reshape(-1, 1)
mu = D.mean(axis=1)
DC = D - vcol(mu)
C = np.dot(DC, DC.T) / (D.shape[1])
var = D.var(1)
std = D.std(1)
print("Mean: ", mu)
print("Covariance matrix: \n", C)
print("Variance: ", var)
print("Standard deviation: ", std)
```
